```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pylab as plt
import matplotlib.patches as mpatches

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

### Data Analysis

I begin by looking at the input features to see how they are distributed and correlate to eachother.

Since the weather is consistent across all files (wc1.csv, wc2.csv, ..., wc3.csv) we can just look at one for correlations.

```python
sns.set_style('whitegrid')

plt.figure(figsize=(18,10))

data = pd.read_csv('apple-watch-heartrate.csv', sep=',', low_memory=False)
heartrate = data.iloc[:,2]                          
sns.violinplot(data=heartrate)
plt.ylabel('Heart Rate')

maximum = mpatches.Patch(label='max   ' + f'{heartrate.max():.2f}')
mean = mpatches.Patch(label='mean ' + f'{heartrate.mean():.2f}')
minimum = mpatches.Patch(label='min    ' + f'{heartrate.min():.2f}')
std = mpatches.Patch(label='std     ' + f'{heartrate.std():.2f}')

plt.legend(handles=[maximum, mean, minimum, std], fontsize='x-large')
plt.show()
```
